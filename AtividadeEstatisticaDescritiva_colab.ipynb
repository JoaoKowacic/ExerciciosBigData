```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as func
import matplotlib.pyplot as plt
```

```python
#informação 1 - Gráfico com as medidas de frequência (Quantidade e Frequência) de acordo com o tipo de pavimento

df.createOrReplaceTempView("dados")

query = """
    SELECT SUM(CAST(paved_road AS INT)) AS paved_road_count,
           SUM(CAST(unpaved_road AS INT)) AS unpaved_road_count
    FROM dados
"""

result = sc.sql(query)

result_pd = result.toPandas()

fig, ax = plt.subplots()
result_pd.T.plot(kind='bar', stacked=True, ax=ax)

ax.set_title('Frequência de Medidas por Tipo de Pavimento')
ax.set_xlabel('Tipo de Pavimento')
ax.set_ylabel('Quantidade de Medidas')
ax.legend(title='Pavimento')

plt.show()
```

```python
#informação 2 - Gráfico com os valores de tendência central (Média, Mediana) da velocidade de acordo com o tipo de pavimento

df.createOrReplaceTempView("dados")

query = """
    SELECT paved_road,
           AVG(CASE WHEN paved_road = 1 THEN speed END) AS media_velocidade_paved,
           PERCENTILE(CASE WHEN paved_road = 1 THEN speed END, 0.5) AS mediana_velocidade_paved,
           AVG(CASE WHEN unpaved_road = 1 THEN speed END) AS media_velocidade_unpaved,
           PERCENTILE(CASE WHEN unpaved_road = 1 THEN speed END, 0.5) AS mediana_velocidade_unpaved
    FROM dados
    GROUP BY paved_road
"""

result = sc.sql(query)

result_pd = result.toPandas()

fig, ax = plt.subplots()

ax.bar(result_pd['paved_road'] - 0.2, result_pd['media_velocidade_paved'], width=0.2, label='Média - Paved')
ax.bar(result_pd['paved_road'], result_pd['mediana_velocidade_paved'], width=0.2, label='Mediana - Paved')

ax.bar(result_pd['paved_road'] + 0.2, result_pd['media_velocidade_unpaved'], width=0.2, label='Média - Unpaved')
ax.bar(result_pd['paved_road'] + 0.4, result_pd['mediana_velocidade_unpaved'], width=0.2, label='Mediana - Unpaved')

ax.set_title('Valores de Tendência Central da Velocidade por Tipo de Pavimento')
ax.set_xlabel('Tipo de Pavimento')
ax.set_ylabel('Velocidade')
ax.legend()

plt.show()
```

```python
#informação 3 - Valores de dispersão (Desvio Padrão, Amplitude) da velocidade de acordo com o tipo de pavimento
#        paved_road X unpaved_road

df.createOrReplaceTempView("dados")

query = """
    SELECT paved_road,
           STDDEV(CASE WHEN paved_road = 1 THEN speed END) AS desvio_padrao_paved,
           MAX(CASE WHEN paved_road = 1 THEN speed END) - MIN(CASE WHEN paved_road = 1 THEN speed END) AS amplitude_paved,
           STDDEV(CASE WHEN unpaved_road = 1 THEN speed END) AS desvio_padrao_unpaved,
           MAX(CASE WHEN unpaved_road = 1 THEN speed END) - MIN(CASE WHEN unpaved_road = 1 THEN speed END) AS amplitude_unpaved
    FROM dados
    GROUP BY paved_road
"""

result = sc.sql(query)

result_pd = result.toPandas()

fig, ax = plt.subplots()


ax.bar(result_pd['paved_road'] - 0.2, result_pd['desvio_padrao_paved'], width=0.2, label='Desvio Padrão - Paved')
ax.bar(result_pd['paved_road'], result_pd['amplitude_paved'], width=0.2, label='Amplitude - Paved')


ax.bar(result_pd['paved_road'] + 0.2, result_pd['desvio_padrao_unpaved'], width=0.2, label='Desvio Padrão - Unpaved')
ax.bar(result_pd['paved_road'] + 0.4, result_pd['amplitude_unpaved'], width=0.2, label='Amplitude - Unpaved')

ax.set_title('Valores de Dispersão da Velocidade por Tipo de Pavimento')
ax.set_xlabel('Tipo de Pavimento')
ax.set_ylabel('Velocidade')
ax.legend()

plt.show()

```

```python
#informação 4 - Valores de dispersão (Percentil Q1, Q2, Q3) da velocidade de acordo com o tipo de pavimento
#        paved_road X unpaved_road
df.createOrReplaceTempView("dados")

query = """
    SELECT paved_road,
           percentile(speed, 0.25) AS q1_paved,
           percentile(speed, 0.5) AS q2_paved,
           percentile(speed, 0.75) AS q3_paved,
           percentile(speed, 0.25) AS q1_unpaved,
           percentile(speed, 0.5) AS q2_unpaved,
           percentile(speed, 0.75) AS q3_unpaved
    FROM dados
    GROUP BY paved_road
"""

result = sc.sql(query)

result_pd = result.toPandas()

fig, ax = plt.subplots()

ax.plot(result_pd['paved_road'] - 0.2, result_pd['q1_paved'], 'o-', label='Q1 - Paved')
ax.plot(result_pd['paved_road'], result_pd['q2_paved'], 'o-', label='Q2 (Mediana) - Paved')
ax.plot(result_pd['paved_road'] + 0.2, result_pd['q3_paved'], 'o-', label='Q3 - Paved')

ax.plot(result_pd['paved_road'] + 0.4, result_pd['q1_unpaved'], 'o-', label='Q1 - Unpaved')
ax.plot(result_pd['paved_road'] + 0.6, result_pd['q2_unpaved'], 'o-', label='Q2 (Mediana) - Unpaved')
ax.plot(result_pd['paved_road'] + 0.8, result_pd['q3_unpaved'], 'o-', label='Q3 - Unpaved')

ax.set_title('Valores de Dispersão (Percentis) da Velocidade por Tipo de Pavimento')
ax.set_xlabel('Tipo de Pavimento')
ax.set_ylabel('Velocidade')
ax.legend()

plt.show()
```

```python
#informação 5 - Boxplot da velocidade de acordo com o pavimento (paved_road ou unpaved)
result = df.select('speed', 'paved_road')\
  .where(df.paved_road == 1 )\
  .select('speed')\
  .summary('mean', '50%', '25%', '75%', 'min', 'max')\
  .collect()

resultUnpaved = df.select('speed', 'unpaved_road')\
  .where(df.unpaved_road == 1 )\
  .select('speed')\
  .summary('mean', '50%', '25%', '75%', 'min', 'max')\
  .collect()
```

```python
meanPaved = float(result[0]['speed'])
medianPaved = float(result[1]['speed'])
q1Paved = float(result[2]['speed'])
q3Paved = float(result[3]['speed'])
minPaved = float(result[4]['speed'])
maxPaved = float(result[5]['speed'])
inter = q3Paved - q1Paved

meanUnpaved = float(resultUnpaved[0]['speed'])
medianUnpaved = float(resultUnpaved[1]['speed'])
q1Unpaved = float(resultUnpaved[2]['speed'])
q3Unpaved = float(resultUnpaved[3]['speed'])
minUnpaved = float(resultUnpaved[4]['speed'])
maxUnpaved = float(resultUnpaved[5]['speed'])
interUnp = q3Unpaved - q1Unpaved
```

```python
stats = [
    {
        'label': 'Paved',
        'med': medianPaved,
        'q1': q1Paved,
        'q3': q3Paved,
        'whislo': max(minPaved, q1Paved - (1.5 * inter)),
        'whishi': min(maxPaved, q3Paved + (1.5 * inter)),
        'fliers': []
    },
    {
        'label': 'Unpaved',
        'med': medianUnpaved,
        'q1': q1Unpaved,
        'q3': q3Unpaved,
        'whislo': max(minUnpaved, q1Unpaved - (1.5 * interUnp)),
        'whishi': min(maxUnpaved, q3Unpaved + (1.5 * interUnp)),
        'fliers': []
    }
]

fig, axes = plt.subplots()
axes.bxp(stats, vert=False)
plt.show()
```

```python
#informação 6 - Boxplot do acelerometro para os eixos X, Y e Z em paved_road

df.createOrReplaceTempView("dados")

query = """
    SELECT speed
    FROM dados
    WHERE paved_road = 1
"""

result = sc.sql(query)

result_pd = result.toPandas()

stats_paved = {
    'label': 'Paved',
    'med': result_pd['speed'].median(),
    'q1': result_pd['speed'].quantile(0.25),
    'q3': result_pd['speed'].quantile(0.75),
    'whislo': result_pd['speed'].min(),
    'whishi': result_pd['speed'].max(),
    'fliers': []
}

outliers_paved = result_pd[(result_pd['speed'] < stats_paved['whislo']) | (result_pd['speed'] > stats_paved['whishi'])]['speed']
stats_paved['fliers'] = outliers_paved.tolist()

stats = [stats_paved]

fig, axes = plt.subplots()
axes.bxp(stats, vert=False)

axes.set_title('Boxplot da Velocidade em Paved Roads')
axes.set_xlabel('Velocidade')

plt.show()
```

```python
#informação 7 - Velocidade (speed) ao longo do tempo (timestamp_gps) em paved_road
#        Atente para a quantidade de entradas
#        Dica, sumarize a velocidade em intervalos de tempo para conseguir exibir
from pyspark.sql.functions import col, from_unixtime, window
from datetime import datetime
import pandas as pd

df.createOrReplaceTempView("dados")

df = df.withColumn("timestamp_gps", from_unixtime("timestamp_gps"))

query = """
    SELECT window(timestamp_gps, '10 minutes').start as time_interval_start,
           AVG(speed) as avg_speed
    FROM dados
    WHERE paved_road = 1
    GROUP BY window(timestamp_gps, '10 minutes')
    ORDER BY time_interval_start
"""

result = sc.sql(query)

result_pd = result.toPandas()

result_pd['time_interval_start'] = pd.to_datetime(result_pd['time_interval_start'])

plt.figure(figsize=(12, 6))

plt.plot(result_pd['time_interval_start'], result_pd['avg_speed'], marker='o', linestyle='-', color='b')

plt.title('Média da Velocidade ao Longo do Tempo em Paved Roads')
plt.xlabel('Tempo (Intervalo de 10 minutos)')
plt.ylabel('Média da Velocidade')
plt.grid(True)

plt.show()
```

```python
#informação 8 - Histograma da velocidade (speed) com 20 intervalos em paved
df.createOrReplaceTempView("dados")

query = """
    SELECT speed
    FROM dados
    WHERE paved_road = 1
"""

result = sc.sql(query)

result_pd = result.toPandas()

plt.figure(figsize=(10, 6))

plt.hist(result_pd['speed'], bins=20, color='blue', edgecolor='black', alpha=0.7)

plt.title('Histograma da Velocidade em Paved Roads')
plt.xlabel('Velocidade')
plt.ylabel('Frequência')
plt.grid(True)

plt.show()
```

```python
#informação 10 - Boxplot da velocidade (speed) em unpaved
df.createOrReplaceTempView("dados")

query = """
    SELECT speed
    FROM dados
    WHERE unpaved_road = 1
"""

result = sc.sql(query)

result_pd = result.toPandas()

median_unpaved = result_pd['speed'].median()
q1_unpaved = result_pd['speed'].quantile(0.25)
q3_unpaved = result_pd['speed'].quantile(0.75)
min_unpaved = result_pd['speed'].min()
max_unpaved = result_pd['speed'].max()
inter_unpaved = q3_unpaved - q1_unpaved

fig, axes = plt.subplots(figsize=(10, 6))

stats_unpaved = {
    'label': 'Unpaved',
    'med': median_unpaved,
    'q1': q1_unpaved,
    'q3': q3_unpaved,
    'whislo': max(min_unpaved, q1_unpaved - (1.5 * inter_unpaved)),
    'whishi': min(max_unpaved, q3_unpaved + (1.5 * inter_unpaved)),
    'fliers': []
}

outliers_unpaved = result_pd[(result_pd['speed'] < stats_unpaved['whislo']) | (result_pd['speed'] > stats_unpaved['whishi'])]['speed']
stats_unpaved['fliers'] = outliers_unpaved.tolist()

axes.bxp([stats_unpaved], vert=False)

axes.set_title('Boxplot da Velocidade em Unpaved Roads')
axes.set_xlabel('Velocidade')

plt.show()
```

```python
#informação 11 - Quantidade de medições de outliers da velocidade (speed) em unpaved

result = df.select('speed', 'paved_road')\
  .where(df.paved_road == 1 )\
  .select('speed')\
  .summary('mean', '50%', '25%', '75%', 'min', 'max')\
  .collect()

meanPaved = float(result[0]['speed'])
medianPaved = float(result[1]['speed'])
q1Paved = float(result[2]['speed'])
q3Paved = float(result[3]['speed'])
minPaved = float(result[4]['speed'])
maxPaved = float(result[5]['speed'])
inter = q3Paved - q1Paved
```

```python

minimoBox = max(minPaved, q1Paved - (1.5 * inter))
maximoBox = min(maxPaved, q3Paved + (1.5 * inter))

resultado = df.select('speed', 'paved_road')\
  .where(df.paved_road == 0)\
  .where((df.speed < minimoBox) | (df.speed > maximoBox))\
  .collect()

outliers = [x['speed'] for x in resultado]

stats = [
    {
        'label': 'Paved',
        'med': medianPaved,
        'q1': q1Paved,
        'q3': q3Paved,
        'whislo': max(minPaved, q1Paved - (1.5 * inter)),
        'whishi': min(maxPaved, q3Paved + (1.5 * inter)),
        'fliers': outliers
    }
]

import matplotlib.pyplot as plt
fig, axes = plt.subplots()
axes.bxp(stats, vert=False)
plt.show()
```
