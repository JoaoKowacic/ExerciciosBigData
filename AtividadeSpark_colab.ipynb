```python
# abre sessão no spark
import pyspark
conf = pyspark.SparkConf()
conf.setMaster('local[*]')
sc = pyspark.SparkContext.getOrCreate(conf)
```

```python
#carrega arquivo da COVID em um rdd
rdd = sc.textFile('file:////content/covid.csv')
```

```python
#informacao 1 Quantidade de pacientes positivos para corona virus (resultadoTeste)
rdd.map(lambda linha: linha.split(';')[11])\
        .filter(lambda test: test == 'Positivo')\
        .count()
```

```python
#informacao 2 Quantidade de pacientes do sexo feminino positivos para corona virus (resultadoTeste)
rdd.map(lambda linha: linha.split(';'))\
        .filter(lambda colunas: colunas[13] == 'Feminino' and colunas[11] == 'Positivo')\
        .count()
```

```python
#informacao 3 Quantidade de pacientes de acordo com o sexo e o resultado do teste (resultadoTeste)

rdd.map(lambda linha: linha.split(';'))\
  .map(lambda colunas: ((colunas[13], colunas[11]), 1))\
  .reduceByKey(lambda x, y: x + y)\
  .collect()


```

```python
#informacao 4 Sintomas mais comuns para casos positivos para corona virus
casos_positivos = rdd.filter(lambda linha: linha.split(';')[11] == 'Positivo')

# mapeia e conta a ocorrência de sintomas
sintomas_contagem = casos_positivos.flatMap(lambda linha: linha.split(';')[4].split(','))\
                                   .map(lambda sintoma: (sintoma.strip(), 1))\
                                   .reduceByKey(lambda x, y: x + y)\
                                   .sortBy(lambda x: x[1], ascending=False)\
                                   .collect()
sintomas_contagem
```

```python
#informacao 5 Sintomas mais comuns para casos não positivos para corona virus
casos_negativos = rdd.filter(lambda linha: linha.split(';')[11] == 'Negativo')

# mapeia e conta a ocorrência de sintomas
sintomas_contagem = casos_negativos.flatMap(lambda linha: linha.split(';')[4].split(','))\
                                   .map(lambda sintoma: (sintoma.strip(), 1))\
                                   .reduceByKey(lambda x, y: x + y)\
                                   .sortBy(lambda x: x[1], ascending=False)\
                                   .collect()
sintomas_contagem
```

```python
#informacao 6 Quantidade de casos positivos no Paraná
rdd.map(lambda linha: linha.split(';'))\
        .filter(lambda colunas: colunas[15] == 'PARANÁ' and colunas[11] == 'Positivo')\
        .count()
```

```python
#informacao 7 Municipio do Paraná com a maior quantidade de óbitos (evolucaoCaso)
casos_parana = rdd.filter(lambda linha: linha.split(';')[15] == 'PARANÁ' and  linha.split(';')[31] == 'Óbito')

obitos_contagem = casos_parana.map(lambda linha: (linha.split(';')[17], 1))\
                                   .reduceByKey(lambda x, y: x + y)\
                                   .collect()
obitos_contagem
```

```python
#informacao 8 Quantidade de municipios no Paraná com casos positivos
casos_positivos = rdd.filter(lambda linha: linha.split(';')[11] == 'Positivo')


municipios_contagem = casos_positivos.map(lambda linha: (linha.split(';')[17], 1))\
                                              .reduceByKey(lambda x, y: x + y)\
                                              .collect()
municipios_contagem
```

```python
#informacao 9 Quantidade de municipios no Paraná sem casos positivos
casos_negativos = rdd.filter(lambda linha: linha.split(';')[11] == 'Negativo')


municipios_contagem = casos_negativos.map(lambda linha: (linha.split(';')[17], 1))\
                                              .reduceByKey(lambda x, y: x + y)\
                                              .collect()
municipios_contagem
```

```python
#informacao 10 Estado com a maior taxa de falecimento de mulheres
casos_parana = rdd.filter(lambda linha: linha.split(';')[13] == 'Feminino' and  linha.split(';')[31] == 'Óbito')

obitos_contagem = casos_parana.map(lambda linha: (linha.split(';')[15], 1))\
                                   .reduceByKey(lambda x, y: x + y)\
                                   .sortBy(lambda x: x[1], ascending=False)\
                                   .collect()
obitos_contagem
```

```python
#informacao 11 Menor idade de mulher positiva para covid
casos = rdd.filter(lambda linha: linha.split(';')[13] == 'Feminino' and  linha.split(';')[11] == 'Positivo')

idade = casos.map(lambda linha: int(linha.split(';')[29]))\
                                .min()

idade
```

```python
#informacao 12 Maior idade de mulher positiva para covid
casos = rdd.filter(lambda linha: linha.split(';')[13] == 'Feminino' and  linha.split(';')[11] == 'Positivo')

idade = casos.map(lambda linha: int(linha.split(';')[29]))\
                                .max()

idade
```

```python
#informacao 13

casos = rdd.filter(lambda linha: linha.split(';')[11] == 'Positivo')

data_casos = casos.map(lambda linha: (linha.split(';')[1], 1))\
                                .reduceByKey(lambda a, b: a + b)\
                                .collect()

data_casos
```

```python
#informacao 14
from datetime import datetime

casos = rdd.filter(lambda linha: linha.split(';')[11] == 'Positivo')

data_casos = casos.map(lambda linha: (datetime.strptime(linha.split(';')[1], '%Y-%m-%dT%H:%M:%S.%f').isocalendar()[1], 1))\
                                .reduceByKey(lambda a, b: a + b)\
                                .collect()

data_casos
```

```python
#informacao 15
casos_parana = rdd.filter(lambda linha: linha.split(';')[31] == 'Óbito')

obitos_contagem = casos_parana.map(lambda linha: (linha.split(';')[29], 1))\
                                   .reduceByKey(lambda x, y: x + y)\
                                   .sortBy(lambda x: x[1], ascending=False)\
                                   .collect()
obitos_contagem
```

```python
#informacao 16
idade_obito = rdd.map(lambda linha: (linha.split(';')[29], linha.split(';')[31]))

obitos = idade_obito.filter(lambda x: x[1] == 'Óbito')

# Conta o número de óbitos por idade
obitos_por_idade = obitos.countByKey()

taxa_obito_idade = {idade: (count / idade_obito.filter(lambda x: x[0] == idade).count()) for idade, count in obitos_por_idade.items()}
```

```python
#informacao 17
rdd_genero_idade_teste = rdd.map(lambda linha: (linha.split(';')[13], linha.split(';')[29], linha.split(';')[11]))


rdd_mulheres_positivas = rdd_genero_idade_teste.filter(lambda x: x[0] == 'Feminino' and x[2] == 'Positivo')


idade_media = rdd_mulheres_positivas.map(lambda x: int(x[1])).mean()
```

```python
#informacao 18
rdd_municipio_genero_teste = rdd.map(lambda linha: (linha.split(';')[17], linha.split(';')[13], linha.split(';')[11]))

# Filtra para casos positivos em mulheres no Paraná
rdd_mulheres_positivas_parana = rdd_municipio_genero_teste.filter(lambda x: x[1] == 'Feminino' and x[2] == 'Positivo' and 'PARANÁ' in x[0])

# Conta o número de casos positivos por município
rdd_casos_por_municipio = rdd_mulheres_positivas_parana.countByKey()

# Encontra o município com a maior quantidade de casos
municipio_maior_casos = max(rdd_casos_por_municipio, key=rdd_casos_por_municipio.get)
```

```python
#informacao 19
from datetime import datetime

# Mapeia a data do teste
rdd_data_teste = rdd.map(lambda linha: linha.split(';')[9])

# Converte as datas para o dia da semana
rdd_dia_semana = rdd_data_teste.map(lambda data: datetime.strptime(data, '%d/%m/%Y').strftime('%A'))

# Conta o número de testes por dia da semana
rdd_testes_por_dia = rdd_dia_semana.countByKey()

# Encontra o dia da semana com a maior quantidade de testes
dia_maior_testes = max(rdd_testes_por_dia, key=rdd_testes_por_dia.get)
```

```python
#informacao 20
from pyspark.sql.functions import date_format

casos_positivos = rdd.filter(lambda linha: linha.split(';')[11] == 'Positivo')

# Mapeia para extrair o dia da semana
casos_por_dia_semana = casos_positivos.map(lambda linha: (linha.split(';')[1], 1)) \
                                      .reduceByKey(lambda x, y: x + y)

# Função para obter o dia da semana
def obter_dia_semana(data):
    try:
        return date_format(data, 'EEEE')
    except:
        return None

# Mapeia para extrair o dia da semana da dataNotificacao
casos_por_dia_semana = casos_por_dia_semana.map(lambda x: (obter_dia_semana(x[0]), x[1])) \
                                           .filter(lambda x: x[0] is not None) \
                                           .reduceByKey(lambda x, y: x + y) \
                                           .sortBy(lambda x: x[1], ascending=False)
```

```python
#informacao 21
# Mapeia o município e o resultado do teste
rdd_municipio_teste = rdd.map(lambda linha: (linha.split(';')[17], linha.split(';')[11]))

# Conta o número de testes por município
rdd_testes_por_municipio = rdd_municipio_teste.countByKey()

# Filtra para municípios com mais de 500 testes
rdd_municipios_500_testes = {municipio: count for municipio, count in rdd_testes_por_municipio.items() if count > 500}

# Filtra para testes que não foram positivos
rdd_testes_nao_positivos = rdd_municipio_teste.filter(lambda x: x[1] != 'Positivo')

# Conta o número de testes não positivos por município
rdd_nao_positivos_por_municipio = rdd_testes_nao_positivos.countByKey()

# Calcula a taxa de testes não positivos por município
taxa_nao_positivos = {municipio: (rdd_nao_positivos_por_municipio[municipio] / count) for municipio, count in rdd_municipios_500_testes.items() if municipio in rdd_nao_positivos_por_municipio}

# Encontra o município com a maior taxa de testes não positivos
municipio_maior_taxa = max(taxa_nao_positivos, key=taxa_nao_positivos.get)
```
